```python
def wrap(text, size=100):
    words = text.split(" ")
    result = ""
    sep = ""
    cur = 0
    for word in words:
        if cur and cur + len(word) > size:
            result += "\n"
            sep = ""
            cur = 0
        add = sep + word
        result += add
        cur += len(add)
        sep = " "
    return result
```
